# Resampling


- Il faut que l'indexe soit en format datetime

```python
df
```

```python
#average CO2-intensity per day

df.resample("d").mean()

# extraire le jour de l'indexe

# GRoup BY jour

# Mean()
```

```python
#max CO2-production per month

df.resample("M").max()[["carbon_intensity_production_avg"]]
```

```python
# total CO2-production and intensity per quarter

df.resample("Q").agg({"carbon_intensity_avg":["min","max","mean","sum"], "carbon_intensity_production_avg" : "median" })
```

# Changing the frequency


```python
#Let's change the frequency to minutes
df.asfreq("min")
```

```python
#change the frequency to half hours
df.asfreq("30min")
```

```python
#we can also use the interpolate function to fill the nans, or any other type of filling with the fillna
df.asfreq("30min").interpolate()
```

## Percentage calculation
